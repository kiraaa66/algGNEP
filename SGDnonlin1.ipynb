#Evaluation of Children

```python
#https://www.youtube.com/watch?v=VVEbwuyoWkU
def gold(func, xl, xu):

  ea = 100
  i = 1

  ratio = (5 ** 0.5 - 1) / 2
  d = ratio * (xu - xl)
  x1 = xl + d
  x2 = xu - d

  f1 = func(x1)
  f2 = func(x2)

  while ea > 0.1:
    if f1 < f2:
      xl = x2
      x2 = x1
      f2 = f1
      x1 = xl + ratio*(xu-xl)
      f1 = func(x1)
    else:
      xu = x1
      x1 = x2
      f1 = f2
      x2 = xu - ratio*(xu-xl)
      f2 = func(x2)

    if f1 < f2:
      xopt = x1
    else:
      xopt = x2

    #print(xopt)
    ea = (1 - ratio) * abs((xu - xl) / xopt) * 100
    i += 1

  return xopt
```

```python
def pen(a,b):
  #https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.golden.html
  def func1(x):
    return x**2 - x*b - x

  def func2(x):
    return x**2 - 0.5*x*a - 2*x

  min1 = gold(func1, 0.00001, 1.0-b)
  min2 = gold(func2, 0.00001, 1.0-a)
  #print(min1)
  shadowX = min1
  shadowY = min2

  pen = math.sqrt((math.pow(a-shadowX,2))+(math.pow(b-shadowY,2)))
  #print(pen)
  return pen
```
